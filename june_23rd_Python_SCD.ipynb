```python
def update_zip_code(df, customer_id, new_zip):
    idx = df[df['customer_id'] == customer_id].index[0]
    current_new = df.at[idx, 'new_zip_code']
    
    if not pd.isna(current_new):
        df.at[idx, 'previous_zip_code'] = current_new
    else:
        df.at[idx, 'previous_zip_code'] = df.at[idx, 'zip_code']
    df.at[idx, 'new_zip_code'] = new_zip

    return df
```

```python
import pandas as pd
from datetime import datetime

def apply_scd_type5(df, customer_history, incoming):

    if df[df['customer_id'] == incoming['customer_id']].empty:
        print("Customer not found")
        return df, customer_history

    current = df[df['customer_id'] == incoming['customer_id']].iloc[0]

    overwrite_fields = ['first_name', 'last_name', 'email', 'city', 'state', 'street', 'phone']

    history_fields = ['zip_code', 'active']

    has_history_change = any(current[field] != incoming[field] for field in history_fields)

    if has_history_change:
        history_entry = current[[
            'customer_id', 'first_name', 'last_name', 'email', 'city',
            'state', 'zip_code', 'active'
        ]].copy()
        history_entry['history_id'] = 1 if customer_history.empty else customer_history['history_id'].max() + 1
        history_entry['change_date'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

        customer_history = pd.concat([customer_history, pd.DataFrame([history_entry])], ignore_index=True)

        df.loc[df['customer_id'] == incoming['customer_id'], 'previous_zip_code'] = current['zip_code']
        df.loc[df['customer_id'] == incoming['customer_id'], 'previous_active'] = current['active']

    for field in overwrite_fields + history_fields:
        if current[field] != incoming[field]:
            df.loc[df['customer_id'] == incoming['customer_id'], field] = incoming[field]

    df.loc[df['customer_id'] == incoming['customer_id'], 'new_zip_code'] = incoming['zip_code']

    return df, customer_history
```

```python

df = pd.DataFrame([{
    'customer_id': 10,
    'first_name': 'Pamelia',
    'last_name': 'Newman',
    'phone': '1234567890',
    'email': 'pamelia.newman@gmail.com',
    'street': '123 Main St',
    'city': 'New York',
    'state': 'NY',
    'zip_code': '10001',
    'active': 'inactive',
    'start_date': '2020-01-01',
    'end_date': '',
    'new_zip_code': '',
    'previous_zip_code': '',
    'previous_active': ''
}])

customer_history = pd.DataFrame(columns=[
    'history_id', 'customer_id', 'first_name', 'last_name', 'email',
    'city', 'state', 'zip_code', 'active', 'change_date'
])

incoming = pd.Series({
    'customer_id': 10,
    'first_name': 'Pamelia',
    'last_name': 'Harsha',
    'email': 'pamelia.newman@gmail.com',
    'city': 'Monroe',
    'state': 'NY',
    'street': '123 Main St',
    'phone': '1234567890',
    'zip_code': '10950',
    'active': 'active'
})

df, customer_history = apply_scd_type5(df, customer_history, incoming)

print("\nUpdated DataFrame (df):")
print(df)

print("\nCustomer History Table:")
print(customer_history)
```
